# 这是Matplotlib的综合练习

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rcParams

rcParams['font.family'] = 'Noto Sans CJK JP'
```

```python
# 导入数据
df = pd.read_csv('../../pandas/data/weather.csv')

# 将日期格式从字符串转换为日期格式
df['date'] = pd.to_datetime(df['date'])
# 取特定时间段的要不图展示出来的数据太多了
df = df[df['date'].dt.year == 2015]
```

```python
# 绘制气温变化趋势图
plt.figure(figsize=(15, 10))

plt.plot(df['date'], df['temp_max'], label='最高气温')
plt.plot(df['date'], df['temp_min'], label='最低气温')
plt.title('2015年气温变化趋势图')
plt.xlabel('日期')
plt.ylabel('气温(℃)')
plt.legend()
plt.show()
```

```python
df['temp_mean'] = (df['temp_max'] + df['temp_min']) / 2
# 绘制气温变化趋势图
plt.figure(figsize=(15, 10))
plt.plot(df['date'], df['temp_mean'], label='平均气温')
plt.title('2015年平均气温变化趋势图')
plt.xlabel('日期')
plt.ylabel('气温(℃)')
plt.legend()
plt.show()
```

```python
# 绘制降水量的直方图
plt.figure(figsize=(15, 10))
plt.hist(df['precipitation'], bins=5, edgecolor='black')
plt.title('2015年降水量直方图')
plt.xlabel('降水量')
plt.ylabel('次数')
plt.show()
```
