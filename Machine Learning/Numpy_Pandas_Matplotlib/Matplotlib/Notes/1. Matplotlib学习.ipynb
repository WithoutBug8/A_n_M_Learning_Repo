````python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import random

rcParams['font.family'] = 'Noto Sans CJK JP'
````

###  1. 绘制折线图
- 看趋势

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 5))
# 绘制图标的数据
month = ['一月', '二月', '三月', '四月', '五月']
sales = [100, 20, 200, 60, 250]
# 绘制折线图
plt.plot(month, sales, label='产品A', color='orange', linestyle='-', linewidth=2, marker='o')
# 添加标题
plt.title('2025年销售趋势', color='Red', fontsize=20)
# 添加坐标轴标签
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额(万元)', fontsize=10)
# 添加图例，左上角的说明
plt.legend(loc='upper left')
# 背景添加网格线
plt.grid(axis='y', alpha=0.2, color='black', linestyle='--')  # 参数True为x和y轴都画网格线 ,alpha控制透明度,linestyle控制线的样式--是虚线
# 设置x y轴刻度的显示大小 rotation让字体旋转角度
plt.xticks(rotation=20, fontsize=10)
plt.yticks(rotation=20, fontsize=10)
# 设置Y轴的范围
plt.ylim(10, 300)
# 每个点进行数据标记
for x, y in zip(month, sales):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10)
# 显示图表
plt.show()
````

### 2. 柱状图
- 看对比

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 5))
subjects = ['语文', '数学', '英语', '科学']
scores = [85, 92, 78, 88]
# 绘制柱状图
plt.bar(subjects, scores, label='Kunkun', width=0.5)
# 添加标题
plt.title('各科成绩', color='Red', fontsize=20)
# 添加坐标轴标签
plt.xlabel('科目', fontsize=10)
plt.ylabel('分数', fontsize=10)
# 添加图例
plt.legend(loc='upper left')
# 背景添加网格线
plt.grid(axis='y', alpha=0.2, color='black', linestyle='--')
# 设置x y轴刻度的显示大小 rotation让字体旋转角度
plt.xticks(rotation=20, fontsize=10)
plt.yticks(rotation=20, fontsize=10)
# 设置Y轴的范围
plt.ylim(0, 100)
# 每个点进行数据标记
for x, y in zip(subjects, scores):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
````

### 3. 条形图
- 看对比

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 5))
countries = ['美国', '中国', '日本', '德国', '澳大利亚']
GDP = [92, 89, 78, 80, 84]
# 绘制条形图
plt.barh(countries, GDP, color='purple')
# 添加标题
plt.title('2025年GDP排名', color='Blue', fontsize=20)
# 添加坐标轴标签
plt.xlabel('GDP', fontsize=10)
plt.ylabel('国家', fontsize=10)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
````

### 4. 饼图
- 看占比,最多不要超过六类要不分辨不出来了

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 5))
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']  # 配色方案
# 绘制饼图
plt.pie(times, labels=things,
        autopct="%.1f%%",  # %.1f%%"表示保留一位小数，%有一个是转义符号
        startangle=90,  # 调整初始画图的角度
        colors=colors
        )
# 添加标题
plt.title('一天之内的时间分配', color='Black', fontsize=20)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
````

### 4.2 圆环图

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 5))
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']  # 配色方案
# 绘制圆环图
plt.pie(times, labels=things,
        autopct="%.1f%%",  # %.1f%%"表示保留一位小数，%有一个是转义符号
        startangle=90,  # 调整初始画图的角度
        colors=colors,
        wedgeprops={'width': 0.6},  # 这里设置圆环图，里面的挖空比例
        pctdistance=0.7  # 这里设置百分比标签离圆心的距离
        )
# 添加标题
plt.title('一天之内的时间分配', color='Black', fontsize=20)
plt.text(0, 0, '总计\n100%', ha='center', va='center', fontsize=10)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
````

### 4.3 爆炸式饼图

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 5))
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']  # 配色方案
explode = [0.1, 0, 0, 0, 0.1]
# 绘制爆炸式饼图
plt.pie(times, labels=things,
        autopct="%.1f%%",  # %.1f%%"表示保留一位小数，%有一个是转义符号
        startangle=0,  # 调整初始画图的角度
        colors=colors,
        explode=explode,  # 用于展示突出的块
        shadow=True  # 突出的块添加阴影
        )
# 添加标题
plt.title('一天之内的时间分配', color='Black', fontsize=20)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
````

### 5. 散点图
- 看数据的聚合离散

````python
# 创建图标，设置大小
plt.figure(figsize=(10, 8))
# 绘制图标的数据
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0, 10)
    x.append(tmp)
    tmp2 = 2 * tmp + random.gauss(0, 2)
    y.append(tmp2)

# 绘制散点图
plt.scatter(x, y,
            color='green',
            alpha=0.5,  # 设置透明度
            s=20,  # 设置原点大小
            label='数据'
            )
# 添加标题
plt.title('X变量与Y变量的关系', color='Red', fontsize=20)
# 添加坐标轴标签
plt.xlabel('X自变量', fontsize=10)
plt.ylabel('Y因变量', fontsize=10)
# 添加图例，左上角的说明
plt.legend(loc='upper left')
# 背景添加网格线
plt.grid(True, alpha=0.2, color='black', linestyle='--')  # 参数True为x和y轴都画网格线 ,alpha控制透明度,linestyle控制线的样式--是虚线
# 设置x y轴刻度的显示大小 rotation让字体旋转角度
plt.xticks(rotation=20, fontsize=10)
plt.yticks(rotation=20, fontsize=10)
# 设置Y轴的范围
plt.ylim(0, 30)
# 在这里可以画一条回归线
plt.plot([0, 10], [0, 20], color='red', linewidth=2)
# 显示图表
plt.show()
````

### 6. 箱线图
用于看数据分布和异常情况
- 上下边缘以外的数值都是**异常值** ，用·表示
- 箱子中的线Q2表示中位数的数据
- 箱子顶部Q3代表的是：从小到大排序，75%位置的数
- 箱子底部Q1代表的是：从小到大排序，25%位置的数
- 箱子的范围IQR就是：25%到75%的数Q3-Q1
- 上下边缘线的范围是：[Q3+1.5*IQR,Q1-1.5*IQR]

````python
# 模拟 3门课的成绩
data = {
    '语文': [82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学': [75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语': [70, 72, 68, 65, 78, 80, 85, 90, 95]
}

plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), tick_labels=data.keys())  # 传入的是数值和对应的名字
plt.title("各科成绩分布(箱线图)")
plt.ylabel("分数")
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
````
