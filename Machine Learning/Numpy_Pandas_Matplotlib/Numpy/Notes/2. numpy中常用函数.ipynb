# Numpy中常用的函数

```python
import numpy as np
```

### 1. 基本的数学函数

```python
# 计算平方根
arr = np.array([4, 9, 16, 25])
print(np.sqrt(arr))
```

```python
# 计算指数,默认以e为底
arr = np.array([1, 2, 3])
print(np.exp(arr))
```

```python
# 计算对数
arr = np.array([1, 10, 100, 1000])
print(np.log(arr))  # log是以e为底
print(np.log10(arr))
print(np.log2(arr))
```

```python
# 计算三角函数
arr = np.array([0, np.pi / 2, np.pi])
print(np.sin(arr))
print(np.cos(arr))
print(np.tan(arr))
```

```python
# 计算统计量
arr = np.array([1, 2, 3, 4, 5])
print(np.abs(arr))
print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))  # 偶数两个数之间的平均值，技术排序后中间的值
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))  # 注意是数组的下标与实际的差距记得+1
print(np.argmax(arr))
```

### 2. 比较函数

```python
# 是否大于、小于、等于
print(np.greater(arr, 3))
print(np.less(arr, 3))
print(np.equal(arr, 3))
# 逻辑与或非
print(np.logical_and([0, 0], [1, 1]))
print(np.logical_or([0, 0], [1, 1]))
print(np.logical_not([0, 0]))
# 检查数组是否有一个为True，是否所有的都为True，自定义条件
print(np.any([0, 0, 0, 0, 0]))
print(np.all([1, 1, 1, 1, 1]))
print(np.where(arr > 3, '及格', '不及格'))  # 条件 满足条件 不满足条件
```

### 3.排序函数

```python
# 排序函数
np.random.seed(42)
arr = np.random.randint(1, 100, 20)
print(arr)
arr.sort()
print(np.argsort(arr))  # 排序后显示索引
print(arr)
```

```python
# 去重函数(去重的同时也会排序)
print(np.unique(arr))
```

```python
# 数组的拼接
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(np.concatenate((arr1, arr2)))  # 拼接
```

```python
# 数组的切割
print(np.split(arr, 5))
```

```python
# 调整数组的形状
print(np.reshape(arr, [4, 5]))
```
