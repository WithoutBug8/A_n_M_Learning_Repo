# 销售数据分析
某产品过去12个月的销售量Series:
- 计算季度平均销量(每3个月为一个季度)
- 找出销量最高的月份
- 计算月环比(和上一个月相比)增长率
- 找出连续增长超过2个月的月份


```python
import numpy as np
import pandas as pd
```

```python
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                  index=pd.date_range('2022-01-01', periods=12, freq='ME'))
# 不用print就可以打印出来
sales
```

```python
#  季度的平均销量
sales.resample('QS').mean()  # 重新采样
```

```python
# 销量最高的月份
print('销量最高的月份是：', sales.idxmax())
```

```python
# 月环比的增长率
print('月环比的增长率\n', sales.pct_change())
```

```python
# 连续增产超过两个月的月份
a = sales.pct_change()
b = a > 0
b[b.rolling(3).sum() == 3].keys().tolist()
```
