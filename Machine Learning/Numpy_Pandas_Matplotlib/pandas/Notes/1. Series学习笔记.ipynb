# 这是pandas中Series的学习笔记

```python
import pandas as pd
```

```python
s = pd.Series([1, 2, 3, 4, 5, 6])
print('注意打印的数据，左边是索引，右边是实际值\n', s)
# 自定义索引index
print('索引是可以自定义的')
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
# 定义name
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='月份')
print(s)
```

```python
# 通过字典创建
s = pd.Series({"a": 1, "b": 2, "c": 3, "d": 4, "e": 5})
print(s)

s1 = pd.Series(s, index=['a', 'c'])
print(s1)
```

### Series常见属性
- Series只是一维的

| 属性 / 方法 | 说明 |
|-------------|------|
| index | Series 的索引对象 |
| values | Series 的值（数组形式） |
| dtype / dtypes | Series 的元素数据类型 |
| shape | Series 的形状 |
| ndim | Series 的维度 |
| size | Series 的元素个数 |
| name | Series 的名称 |
| loc[] | 显式索引：按标签索引或切片 |
| iloc[] | 隐式索引：按位置索引或切片 |
| at[] | 使用标签访问单个元素 |
| iat[] | 使用位置访问单个元素 |

```python
# Series属性
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='尚硅谷')
print('索引', s.index)
print('数值', s.values)
print('形状', s.shape)
print('维度', s.ndim)
print('名称', s.name)
print('数据类型', s.dtype)

# 和索引相关
print('显式索引', s.loc['a'])
print('显式索引,支持切片\n', s.loc['a':'c'])
print('隐式索引', s.iloc[0])

print('直接获取单个的数值(显式)，不支持切片', s.at['a'])
print('直接获取单个的数值(隐式)', s.iat[0])
```
